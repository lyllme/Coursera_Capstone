## importing data from url

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source, 'lxml')

csv_file = open('canada3.csv', 'w', newline = "")

csv_writer = csv.writer(csv_file)
csv_writer.writerow(['PostalCode', 'Borough', 'Neighborhood'])

table = soup.findAll("table")[0]
```

## taking data/export to csv

```python
size = len(table.find_all("b"))

import types

for i in range(size):
    line = table.findAll("p")[i]

    # PostalCode
    postal = line.find("b")
    PostalCode = postal.text
    print(PostalCode)
    
    # Borough
    spanB = line.find("span").text
    spanC = spanB.replace('(', '/')
    Borough = spanC.split('/')[0]
    if i == 114:
        Borough = ['Mississauga','Canada Post','Gateway Processing Centre']
    if i == 152:
        Borough = ['Etobicoke','Northwest']
    if i == 168:
        Borough = ['East Toronto', 'Business reply mail', 'Processing Centre', '969 Eastern']
        
    if type(Borough) is list:
        Borough = ', '.join(Borough)
        
    print(Borough)
    
    # Neighborhood
    spanA = line.find("span").text
    spanB = spanA.replace('(', '/').replace('/', ',').replace(')', '')
    spanC = [x.strip() for x in spanB.split(',')]
    neighborhood = spanC[1:]
    
    if type(neighborhood) is list:
        Neighborhood = ', '.join(neighborhood)
        
    print(Neighborhood)
    
    csv_writer.writerow([PostalCode, Borough, Neighborhood])
    
csv_file.close()
```
