```python
import pandas as pd
df = pd.read_csv('canada3.csv')
```

```python
# deleting NaN cells, and count the remainings
df_noNaN = df.dropna(how='any', axis = 0)
df_noNaN.shape
```

```python
df2 = pd.read_csv('Geospatial_Coordinates.csv')
df2 = df2.rename(columns={'Postal Code': 'PostalCode'})
df_incidents = pd.merge(df_noNaN, df2, on='PostalCode', how='inner')
df_incidents
```

```python
#!pip install folium==0.10.1
import folium
print(folium.__version__)
```

```python
# define the world map
world_map = folium.Map()

### type your answer here
tronto_latitude = 43.653226  
tronto_longitude = -79.383184 
# define the world map centered around Canada with a low zoom level
world_map = folium.Map(location=[tronto_latitude, tronto_longitude], zoom_start=10)

limit = 200
df_incidents = df_incidents.iloc[0:limit, :]

# display world map
world_map
```

```python
df_incidents.shape
```

```python
l = ["a", "b"]
l_replace = [s.replace("b", "a") for s in l]
l_replace
```

```python
changerow = []
for post in df_incidents.PostalCode:
    changerow.append('\n')
    
l_borough = []

l_borough = [s.replace(",", ",\n") for s in df_incidents.Borough]

    
l_popinfo = []
for post, changerow, borough in zip(df_incidents.PostalCode, changerow, l_borough):
    l_popinfo.append(''.join(post+changerow+borough))
    #print(l_popinfo)


    
#l_popinfo
# to dataframe
df_incidents['popinfo']= l_popinfo
df_incidents
```

# answer

```python
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
#for lat, lng, label in zip(df_incidents.Latitude, df_incidents.Longitude, df_incidents.PostalCode):
for lat, lng, label in zip(df_incidents.Latitude, df_incidents.Longitude, df_incidents.popinfo):   
    # incidents.add_child(
    folium.Marker(
    location = [lat, lng],
    icon = None,
    popup=label,
    ).add_to(incidents)
        
# add incidents to map
world_map.add_child(incidents)
```
